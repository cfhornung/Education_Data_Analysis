### I hypothesis there is a correlation between the unemployment total and labor force total for countries around the world. There will be a inverse (negative) correlation between these two indicatiors where as labor force is high, unemployment will be low and vice versa. 

### Importing Pandas, reading my CSV file, and declaring the named constants for my two indicators.

```python
import pandas as pd
education = pd.read_csv("../../data/databank_education_2018.csv")

UN_EMPLOYMENT = education["SL.UEM.TOTL.ZS"]
LABOR_FORCE = education["SL.TLF.TOTL.IN"]
COUNTRY =  education["Country"]
COUNTRY
```

### Taking a look at all the data.

```python
education = pd.DataFrame(education)
education
```

### Creating a new DataFrame with only the Country, Unemployment Percentages, and Labor Force Totals.

```python
edu = education[["Country", "SL.UEM.TOTL.ZS", "SL.TLF.TOTL.IN"]].copy()
edu.head
```

### Renaming the columns so it is more readable and usable instead of using named constants for table and presentation purposes.

```python
education = edu.rename(columns={"SL.UEM.TOTL.ZS": "Unemployment_Percent", "SL.TLF.TOTL.IN": "LaborForce_Total"})
```

### Taking out the Nan rows for each column.

```python
education = education[education.Country.notna() & education.Unemployment_Percent.notna() & education.LaborForce_Total.notna()]
```

### Seeing that the rows went down from 244 to 225 with the Nan Values taken out.

```python
education.shape
```

### Taking a look at my new DataFrame.

```python
education.head(10)
```

### These are the highest unemployment rates with the corresponding labor force totals for each Country.

```python
highest_unemployment = education.sort_values(by='Unemployment_Percent', ascending=False)
highest_unemployment.head(10)
```

### These are the highest labor force totals rates with the corresponding unemployment rates for each Country.

```python
highest_laborforce = education.sort_values(by='LaborForce_Total', ascending=False)
highest_laborforce.head(10)
```

### Creating a scatter plot with unemployment rates being the x value and labor force totals being the y values. 

```python
scatter = education.plot.scatter(x = 'Unemployment_Percent', y = 'LaborForce_Total', title = "Unemployment and Labor Correlation")
scatter
```

### Importing some libraries to do some further analysis.

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

### Setting the x and y values to calculate correlation.

```python
x = education.Unemployment_Percent
y = education.LaborForce_Total
```

### Looking at the correlation between Unemployment rates and labor force totals.

```python
r = np.corrcoef(x, y)
r
```

### Looking at Pearsons stats.

```python
scipy.stats.pearsonr(x, y)
```

### Looking at Spearmans stats.

```python
scipy.stats.spearmanr(x, y)
```

### Looking at Kendalls stats.

```python
scipy.stats.kendalltau(x, y)
```

### Creating a scatter plot with matplotlib with a linear regression line.

```python
plt.plot(x, y, 'o')
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)
```

# Conclusion
## My hypothesis was right! ... Kinda.
### After looking at the p values and correlations, there is a very weak negative correlation between unemployment rates and labor force totals. This might be because they are two different units of measurments. One is a percentage and the other is totals. If I had the same data with the same units of measurements, I might see a larger negative correlation (a strong correlation) between these two indicators.
